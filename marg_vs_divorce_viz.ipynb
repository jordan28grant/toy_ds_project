The question of interest is if the margarine comsumption is related to divorce rates in the state of Maine, US. From the graph we see that the trends of margarine consumption and divorce rates per time appear to parralel one another. Both decrease as time progresses and therefore they both have a negative relationship with time. This denotes a correlation between margarine consumption and divorce rates but it does not infer causation. Firstly, it is unclear which variable came first, less margarine consumption or lower divorce rates. Secondly, from the data gathered other confounding variables cannot be ruled out.
